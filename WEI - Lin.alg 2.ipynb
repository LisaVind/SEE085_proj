```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data= pd.read_csv("/Users/edvineng/Documents/Vinylcrackle/matte 2/CHALMERS/Programering/vattenuttag.csv", sep=";")
lista= np.array(data)

xlist=[]
ylist=[]

for i in range(9):
    x=lista[i, 0]
    y=lista[i, 1]
    xlist.append(x)
    ylist.append(y)


xx= np.linspace(1975, 2015)
curvefit= np.polyfit(xlist, ylist, 4)
yfunklist= np.polyval(curvefit, xx)


#fig = plt.figure()
#ax = fig.add_subplot(1, 1, 1)
#ax.set_facecolor('xkcd:cyan')

plt.plot(xlist,ylist, ".", label="Dokumenterade vrden")
plt.plot(xx, yfunklist, label="Kurvanpassning, 4:e gradare")
plt.ylabel("Vattenuttag [$m^3$]")
plt.xlabel("Tid [r]")
plt.legend()
plt.show()


#%%
ddata= pd.read_csv("/Users/edvineng/Documents/Vinylcrackle/matte 2/CHALMERS/Linjr Algebra/tillgngligt_vatten.csv", sep=";")
llista=np.array(ddata)

xxlist=[]

yylist=[1.71368E+11, 1.71045E+11, 1.71336E+11, 1.70972E+11, 1.69161E+11, 1.71102E+11, 1.70591E+11, 1.71377E+11, 1.71372E+11, 1.71234E+11, 1.74718E+11]

for i in range(11):
    xx=llista[i, 0]
    xxlist.append(xx)
 
xxx= np.linspace(1967, 2017)
curvefit= np.polyfit(xxlist, yylist, 4)
yyfunklist= np.polyval(curvefit, xxx)    

plt.plot(xxlist, yylist, "o", label="Dokumenterade vrden")
plt.plot(xxx, yyfunklist,label="Kurvanpassning, 4:e gradare" )
plt.ylabel("Tillngligt vatten [$m^3$]")
plt.xlabel("Tid [r]")
plt.legend()
plt.show()

#%%
wei= yfunklist/yyfunklist
weiprocent= wei*100
plt.plot(xxx, weiprocent, label="WEI")
plt.ylabel("Vattenuttag/ Tillngligt vatten [%]")
plt.xlabel("Tid [r]")
plt.legend()
plt.show()


#%%
ymedlist=[]
ymedlist_2=[]
ymedlist_3=[]
ymedlist_4=[]
ymedlist_5=[]

ymedvalist=[]


    
for i in range(0,9):
    ymed=wei[i]
    ymedlist.append(ymed)    
for i in range(10,19):
    ymed=wei[i]
    ymedlist_2.append(ymed)
for i in range(20,29):
    ymed=wei[i]
    ymedlist_3.append(ymed)
for i in range(30,39):
    ymed=wei[i]
    ymedlist_4.append(ymed)
for i in range(40,49):
    ymed=wei[i]
    ymedlist_5.append(ymed)

    
ymedval=sum(ymedlist)/10
ymedval_2=sum(ymedlist_2)/10
ymedval_3= sum(ymedlist_3)/10
ymedval_4=sum(ymedlist_4)/10
ymedval_5=sum(ymedlist_5)/10

ymedvalist.append(ymedval)
ymedvalist.append(ymedval_2)
ymedvalist.append(ymedval_3)
ymedvalist.append(ymedval_4)
ymedvalist.append(ymedval_5)

xxxx=np.linspace(1967, 2015, 5)


ymedvalistprocent=np.array(ymedvalist)*100

plt.plot(xxxx, ymedvalistprocent, label="Frdnring i medelvrdet (WEI)")
plt.ylabel("Vattenuttag/ Tillngligt vatten [%]")
plt.xlabel("Tid [r]")
plt.legend()
plt.show()
```
